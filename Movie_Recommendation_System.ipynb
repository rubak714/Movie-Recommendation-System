```python
#importing dependencies
import numpy as np
import pandas as pd
import difflib
#to convert text data into numerical values
from sklearn.feature_extraction.text import TfidfVectorizer
#to find simiarities among movies to get closer match to input movie
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre-processing


```python
#loading the data from the csv file to a pandas dataframe
movies_data=pd.read_csv('/content/movies.csv')
```

```python
#printing the first five rows of the data frame
movies_data.head()
```

```python
#number of rows and columns in the data frame
movies_data.shape
```

```python
#selecting relevant features for recommendation(content based)
selected_features=['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string
#The fillna() method replaces the null values
#feature is 'genres' as index=0,then feature will take 'keywords' as index=1 and so on
for feature in selected_features:
   movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all the five selected features
combined_features=movies_data['genres']+movies_data['keywords']+movies_data['tagline']+movies_data['cast']+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting the text data into features 'vectors'
#Word Embeddings or Word vectorization is a methodology in NLP to map words or phrases from vocabulary 
#to a corresponding vector of real numbers which used to find word predictions, word similarities/semantics. 
#The process of converting words into numbers are called Vectorization.

vectorizer=TfidfVectorizer()
```

```python
#The fit method is calculating the mean and variance of each of the features present in our data,
#The transform method is transforming all the features using the respective mean and variance.
feature_vectors=vectorizer.fit_transform(combined_features)
#feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

Cosine Similarity


```python
#achieving the similarity score using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

Getting the movie name from user

```python
#getting the movie name from the user
my_movie_name=input("Enter your favourite movie name: ")
```

```python
#creating a list with all the movie names given in the dataset
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user
#get_close_matches() is a function that is available in the difflib Python package. 
#The difflib module provides classes and functions for comparing sequences. 
#We can use this module to compare files and produce information about file differences in various formats.
find_close_match=difflib.get_close_matches(my_movie_name,list_of_all_titles)
```

```python
print(find_close_match)
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
#finding the index of the movie with this title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
```

```python
# getting a list of similar movies

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```

Movie Recommendation Sytem

```python
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')


i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
